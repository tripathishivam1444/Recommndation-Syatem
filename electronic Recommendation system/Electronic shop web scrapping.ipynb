## Main Full Final Code

```python
# ANSI escape code for bold text
bold_text = "\033[1m"

# Reset the text formatting to normal after the bold text
reset_text = "\033[0m"

# Print text in bold format
print(bold_text + "ram" + reset_text)
```

```python
y = 1
for name ,link in zip(df['Name'][10:], df['Link'][10:]):
    print(bold_text + name+ reset_text)
    number = y, "/", df.shape[0]
    y = y+1
```

```python
Product_name = []
Details = []
Price = []
Image_link = []
Discription_link = []
Product_discription = []
```

```python
w = 1
y = 1

for name, link in zip(df['Name'][10:], df['Link'][10:]):
    number = str(y) + "/" + str(df.shape[0])
    print(bold_text + name + reset_text + "\t" + number)
    y = y + 1

    httpobject = urlopen(link)
    webdata1 = httpobject.read()
    soupdata = soup(webdata1)
    s = 10
    for i in soupdata.findAll('div', {'class': "product-thumb"}):
        
        print(w)
        print(f"\n {name}-------- {number} --- {s} ")
        print("\t\t", '  Name ----> ', i.find('div', class_="caption").find('a').text)  # <------------Name
        Product_name.append(i.find('div', class_="caption").find('a').text)

        # print("\t\t", 'Details ----> ', i.find('div', class_="caption").find('p').text)  # <------------ detail
        Details.append(i.find('div', class_="caption").find('p').text)

        # print("\t\t", 'Price --->', i.find('div', class_='caption').find('p', class_='price').text)  # <------- Price
        Price.append(i.find('div', class_='caption').find('p', class_='price').text)

        # print("\t\t", 'Image link----> ', i.find('div', class_="image").find('img')['src'])  # <------------ product image link
        Image_link.append(i.find('div', class_="image").find('img')['src'])

        # print("\t\t", 'Discription link --->', i.find('div', class_='caption').find('a')['href'])  # <------- Deep Description Link
        Discription_link.append(i.find('div', class_='caption').find('a')['href'])

        try:
            httpobject2 = urlopen(i.find('div', class_='caption').find('a')['href'])
            webdata2 = httpobject2.read()
            for j in soup(webdata2).findAll('div', class_='tab-pane active'):
                # print("\t\t", 'Product description --->', j.text)  # <------ Deep Description
                Product_discription.append(j.text)
        except Exception as e:
            print("\t\t Error fetching product description:", e)
            Product_discription.append(np.nan)

        # print(f" =--------------{name}------  {number} --- {s} \n")
        w = w + 1
        s = s + 1
```
